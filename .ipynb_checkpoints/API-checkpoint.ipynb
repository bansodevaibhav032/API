```python
import pandas as pd 

def get_user():
    info=[]

    for user in RandomUser.generate_users(20):
        info.append({"Name":user.get_full_name(),
                    "Gender":user.get_gender(),
                    "City":user.get_city(),
                    "State":user.get_state(),
                    "Email":user.get_email(),
                    "DOB":user.get_dob(),
                    "Pictures":user.get_picture()                    
        })
    return pd.DataFrame(info)     

get_user()
df1=pd.DataFrame(get_user()) 
df1
```
